```python
#Currently, we have all the sentences of all the documents as single list. 
#We need to have a list of lists where each element represents an article in the dump. 


document_data = [] #List to store elements where each element is a list representing texts in an article
individual_data = [] #List to store texts of each artile
```

```python
#This is the string which seperates each article. 
#We will check for the occurance of this string and will populate our list. 

check_value = '#Article: '
```

```python
#Here we check for the 'Article' by comparing each element with the 'check_value'. We first extract the contents of individual article. 
#When we enconter the term '#Article: ', we append this value as a single list in the 'document_data'. 
#In this way, our 'docment_data' will have list of lists where each individual element represents contents of individual article. 

for value in clean_data:
    if check_value not in value:
        individual_data.append(value)
    else:
        document_data.append(individual_data)
        individual_data = []
        individual_data.append(value)

document_data.append(individual_data)
```

```python
#Let's extract the titles of each article. It is present as the first element in each list that populates 'document_data'. 
#We also remove the title (#Article) from the contents of each article. 

titles = []

for idx in range(0,len(document_data)):
    titles.append(document_data[idx][0])
    document_data[idx].pop(0)
    
```
